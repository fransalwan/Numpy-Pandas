```python
df_restaurants = pd.read_csv('Top250.csv')
df_restaurants
```

```python
df_restaurants.info()
```

```python
# Soal 1
# Replace / create a new column that converts YOY_Sales column into decimal format 
# (e.g. 4.9% (String)--> 0.049(Float), 13.0%-->0.13)
```

```python
# float(df_restaurants['YOY_Sales'][0].replace('%', '')) / 100

# define a function, then apply it to every row
def convert_to_decimal(x):
    return float(x.replace('%', '')) / 100
    
df_restaurants['YOY_Sales'] =  df_restaurants['YOY_Sales'].apply(convert_to_decimal)
# define a lambda function, idem
```

```python
df_restaurants
```

```python
# Soal 2
# Retrieve the name of Quick Service & Mexican restaurants that are in rank 1-50
```

```python
df_restaurants[
    (df_restaurants['Segment_Category'] == 'Quick Service & Mexican') 
    & (df_restaurants['Rank'].between(1, 50))
]
```

```python
# Soal 3
# Retrieve the top 20 Quick Service & Burger Restaurants based on their YOY Sales.
```

```python
df_restaurants[
    df_restaurants['Segment_Category'] == 'Quick Service & Burger'
].sort_values(by = 'YOY_Sales', ascending = False).iloc[:20]
```

```python
# Soal 4
# Calculate the average Sales in each restaurant category
```

```python
df_restaurants.groupby('Headquarters')['Sales'].mean()
```
